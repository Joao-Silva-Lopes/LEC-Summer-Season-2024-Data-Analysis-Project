## Let's analyze the 10 most attacked and banned champions during the championship

```python
df_porcentagem_de_campeao = ['Ban', 'Ban_Opponent', 'Pick', 'Pick_Opponent','Outcome']
df_porcentagem_de_campeao = df[df_porcentagem_de_campeao]

# Count the frequency of each champion in the 'Pick' column

top_picks = df_porcentagem_de_campeao['Pick'].value_counts().head(10)
top_picks_opponent = df_porcentagem_de_campeao['Pick_Opponent'].value_counts().head(10)

# Count the frequency of each champion in the 'Ban' column

top_bans = df_porcentagem_de_campeao['Ban'].value_counts().head(10)
top_bans_opponent = df_porcentagem_de_campeao['Ban_Opponent'].value_counts().head(10)

# Combine picks and bans

top_picks_combined = pd.concat([top_picks, top_picks_opponent], axis=1, keys=['Pick', 'Pick_Opponent']).fillna(0).sum(axis=1).sort_values(ascending=False).head(10)
top_bans_combined = pd.concat([top_bans, top_bans_opponent], axis=1, keys=['Ban', 'Ban_Opponent']).fillna(0).sum(axis=1).sort_values(ascending=False).head(10)

# Add win and loss columns for picks

top_picks_combined = top_picks_combined.to_frame(name='Total_Picks')
top_picks_combined['Wins'] = 0
top_picks_combined['Losses'] = 0

# Count wins and losses for each champion in the picks

for champ in top_picks_combined.index:
    top_picks_combined.at[champ, 'Wins'] = (
        df_porcentagem_de_campeao[(df_porcentagem_de_campeao['Pick'] == champ) & (df_porcentagem_de_campeao['Outcome'] == 'Win')].shape[0] +
        df_porcentagem_de_campeao[(df_porcentagem_de_campeao['Pick_Opponent'] == champ) & (df_porcentagem_de_campeao['Outcome'] == 'Loss')].shape[0]
    )
    top_picks_combined.at[champ, 'Losses'] = (
        df_porcentagem_de_campeao[(df_porcentagem_de_campeao['Pick'] == champ) & (df_porcentagem_de_campeao['Outcome'] == 'Loss')].shape[0] +
        df_porcentagem_de_campeao[(df_porcentagem_de_campeao['Pick_Opponent'] == champ) & (df_porcentagem_de_campeao['Outcome'] == 'Win')].shape[0]
    )

# Calculate winning percentage for picks

top_picks_combined['Win_Percentage'] = (top_picks_combined['Wins'] / top_picks_combined['Total_Picks']) * 100


# Format win percentage with two decimal places for picks

top_picks_combined['Win_Percentage'] = top_picks_combined['Win_Percentage'].map(lambda x: f'{x:.2f}%')

# Add win and loss columns for bans

top_bans_combined = top_bans_combined.to_frame(name='Total_Bans')
top_bans_combined['Wins'] = 0
top_bans_combined['Losses'] = 0

# Count wins and losses for each champion in bans

for champ in top_bans_combined.index:
    top_bans_combined.at[champ, 'Wins'] = (
        df_porcentagem_de_campeao[(df_porcentagem_de_campeao['Ban'] == champ) & (df_porcentagem_de_campeao['Outcome'] == 'Win')].shape[0] +
        df_porcentagem_de_campeao[(df_porcentagem_de_campeao['Ban_Opponent'] == champ) & (df_porcentagem_de_campeao['Outcome'] == 'Loss')].shape[0]
    )
    top_bans_combined.at[champ, 'Losses'] = (
        df_porcentagem_de_campeao[(df_porcentagem_de_campeao['Ban'] == champ) & (df_porcentagem_de_campeao['Outcome'] == 'Loss')].shape[0] +
        df_porcentagem_de_campeao[(df_porcentagem_de_campeao['Ban_Opponent'] == champ) & (df_porcentagem_de_campeao['Outcome'] == 'Win')].shape[0]
    )

# Calcular porcentagem de vitÃ³rias para bans
top_bans_combined['Win_Percentage'] = (top_bans_combined['Wins'] / top_bans_combined['Total_Bans']) * 100

# Formatar porcentagem de vitÃ³rias com duas casas decimais para bans
top_bans_combined['Win_Percentage'] = top_bans_combined['Win_Percentage'].map(lambda x: f'{x:.2f}%')

```

### Which will be the 5 teams with the most wins and the 5 teams with the most losses

```python
df_melhores_times = df[['Team', 'Opponent_Team', 'Outcome']]

# Count the number of wins and losses for each team
win_counts = df_melhores_times[df_melhores_times['Outcome'] == 'Win']['Team'].value_counts()
loss_counts = df_melhores_times[df_melhores_times['Outcome'] == 'Loss']['Team'].value_counts()

# Select the 5 teams with the most wins and the 5 teams with the most losses
top_5_wins = win_counts.head(5)
top_5_losses = loss_counts.head(5)

# Create DataFrames for easier viewing
top_5_wins_df = pd.DataFrame({'Team': top_5_wins.index, 'Wins': top_5_wins.values})
top_5_losses_df = pd.DataFrame({'Team': top_5_losses.index, 'Losses': top_5_losses.values})

# Configure the chart
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Chart of teams with the most wins
bars_wins = ax[0].bar(top_5_wins_df['Team'], top_5_wins_df['Wins'], color='green')
ax[0].set_title('Top 5 Teams with the Most Wins')
ax[0].set_xlabel('Teams')
ax[0].set_ylabel('Number of Wins')

# Add the values â€‹â€‹on top of the win bars
for bar in bars_wins:
    yval = bar.get_height()
    ax[0].text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom')  # va='bottom' coloca o texto acima da barra

# Chart of teams with the most defeats
bars_losses = ax[1].bar(top_5_losses_df['Team'], top_5_losses_df['Losses'], color='red')
ax[1].set_title('Top 5 Teams with the Most Defeats')
ax[1].set_xlabel('Teams')
ax[1].set_ylabel('Number of Defeats')

# Add the values â€‹â€‹on top of the defeat bars
for bar in bars_losses:
    yval = bar.get_height()
    ax[1].text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom')  # va='bottom' coloca o texto acima da barra

# Adjust layout
plt.tight_layout()

# Display the graph
plt.show()
```


#### We can see that SK and BDS are the winning team, they will probably face each other in future matches


#### We can also see that RGE and VIT are the teams with the most defeats, they will probably face each other

## And which teams would have the most map vision?

```python
df_visao_de_jogo = df[['Player', 'Role', 'Team', 'Opponent_Team', 'Opponent_Player', 'Wards placed',
       'Wards destroyed', 'Control Wards Purchased', 'Detector Wards Placed']]

# Group and add metrics for each team
wards_placed = df_visao_de_jogo.groupby('Team')['Wards placed'].sum()
wards_destroyed = df_visao_de_jogo.groupby('Team')['Wards destroyed'].sum()
control_wards_purchased = df_visao_de_jogo.groupby('Team')['Control Wards Purchased'].sum()

# Select the 3 teams with the most values â€‹â€‹for each metric
top_3_wards_placed = wards_placed.sort_values(ascending=False).head(3)
top_3_wards_destroyed = wards_destroyed.sort_values(ascending=False).head(3)
top_3_control_wards_purchased = control_wards_purchased.sort_values(ascending=False).head(3)

# Create DataFrames for easier viewing
top_3_wards_placed_df = top_3_wards_placed.reset_index()
top_3_wards_destroyed_df = top_3_wards_destroyed.reset_index()
top_3_control_wards_purchased_df = top_3_control_wards_purchased.reset_index()

# Configure the chart
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Wards Placed Chart
bars_wards_placed = ax[0].bar(top_3_wards_placed_df['Team'], top_3_wards_placed_df['Wards placed'], color='blue')
ax[0].set_title('Top 3 Teams with Most Wards Placed')
ax[0].set_xlabel('Teams')
ax[0].set_ylabel('Number of Wards Placed')

# Add the values â€‹â€‹on top of the wards placed bars
for bar in bars_wards_placed:
    yval = bar.get_height()
    ax[0].text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom')

# Destroyed Wards Chart
bars_wards_destroyed = ax[1].bar(top_3_wards_destroyed_df['Team'], top_3_wards_destroyed_df['Wards destroyed'], color='orange')
ax[1].set_title('Top 3 Teams with the Most Wards Destroyed')
ax[1].set_xlabel('Teams')
ax[1].set_ylabel('Number of Wards Destroyed')

# Add the values â€‹â€‹on top of the destroyed wards bars
for bar in bars_wards_destroyed:
    yval = bar.get_height()
    ax[1].text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom')

# Control Wards Purchased Chart
bars_control_wards_purchased = ax[2].bar(top_3_control_wards_purchased_df['Team'], top_3_control_wards_purchased_df['Control Wards Purchased'], color='green')
ax[2].set_title('Top 3 Teams with Most Control Wards Purchased')
ax[2].set_xlabel('Teams')
ax[2].set_ylabel('Number of Control Wards Purchased')


# Add the values â€‹â€‹above the control wards purchased bars
for bar in bars_control_wards_purchased:
    yval = bar.get_height()
    ax[2].text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom')

# Adjust layout
plt.tight_layout()

# Display the graph
plt.show()
```


#### Looking at this graph you can see that the MDK team has a lot of vision control, but why is it one of the teams that loses the most? It's an interesting question, since game vision is one of the fundamental parts of being a great player.
